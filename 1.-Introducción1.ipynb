```python
#ejemplo 1
import tensorflow as tf

w = tf.add(4,5)
print w
```

```python
sess = tf.Session()
print sess.run(w)
sess.close()
```

```python
with tf.Session() as sess:
    print sess.run(w)
```

```python
import tensorflow as tf
#ejemplo 2
x = 2
y = 1
op1 = tf.add(x, y)
op2 = tf.multiply(x, y)
op3 = tf.pow(op2, op1)
with tf.Session() as sess:
    writer = tf.summary.FileWriter('./graphs', sess.graph)
    op3 = sess.run(op3)
print op3
```

```python
import tensorflow as tf
x = tf.add(2, 3)
with tf.Session() as sess:
    #agregar esta linea para usar tensorboard
    writer = tf.summary.FileWriter('./graph', sess.graph)
    print sess.run(x)
writer.close() 
```
